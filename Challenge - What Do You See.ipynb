```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('day.csv')
print(df)
```

```python
plt.scatter(x = df['temp'], y = df['cnt'])
plt.title('')
plt.show()
```

```python
plt.boxplot(df['windspeed'])
plt.title('Normalized Windspeed')
plt.show()
```

```python
plt.hist(df['cnt'])
plt.title('Rentals')
plt.xlabel('Counts of Rentals')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.plot(df['casual'])
plt.plot(df['registered'])
plt.ylabel('Frequency')
plt.title('Casual vs. Registered')
plt.legend()
plt.show()
```
